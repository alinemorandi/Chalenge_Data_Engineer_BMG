```python
from pyspark.sql import functions as F
from pyspark.sql.types import IntegerType, DateType

# rea bronze
df_bronze = spark.table("workspace.bronze.beneficiarios_ans")

# apply transformations
df_silver = df_bronze.select(
    F.col("ID_CMPT_MOVEL").cast("string").alias("competencia"),
    F.col("CD_OPERADORA").cast("int").alias("id_operadora"),
    F.trim(F.col("NM_RAZAO_SOCIAL")).alias("nome_operadora"),
    F.col("CD_MUNICIPIO").cast("int").alias("id_municipio"),
    F.trim(F.col("NM_MUNICIPIO")).alias("nome_municipio"),
    F.trim(F.col("SG_UF")).alias("uf"),
    F.trim(F.col("DE_FAIXA_ETARIA")).alias("faixa_etaria"),
    F.trim(F.col("DE_SEGMENTACAO_PLANO")).alias("segmentacao_plano"),
    F.col("QT_BENEFICIARIO_ATIVO").cast("int").alias("qtd_ativos"),
    F.to_date(F.col("DT_CARGA")).alias("data_carga")
)

# save delta table
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.silver")
df_silver.write.format("delta").mode("overwrite").saveAsTable("workspace.silver.beneficiarios_ans")

print("Silver ok.")
```
