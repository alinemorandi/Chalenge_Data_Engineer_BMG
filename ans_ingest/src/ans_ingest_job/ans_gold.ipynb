```python
# read silver
import pyspark.sql.functions as F
df_silver = spark.table("workspace.silver.beneficiarios_ans")


# create aggregated table
df_gold = df_silver.groupBy(
    "nome_operadora", 
    "faixa_etaria", 
    "nome_municipio", 
    "uf"
).agg(
    F.sum("qtd_ativos").alias("total_beneficiarios_ativos")
)

# save gold
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.gold")
df_gold.write.format("delta").mode("overwrite").saveAsTable("workspace.gold.agg_beneficiarios_ans")

print("Gold is ready.")
```
